```python
from enum import Enum
import time
```

```python
class ticket_stat(Enum):
    active=1
    complete=2
class vehicletype (Enum) :
    four_vehiler = 1
    two_vehiler = 2
class spottype (Enum) :
    free = 1
    occupied = 2
class parkinglot():
    def __init__(self,name,address):
        self.name=name
        self.address=address
        self.level=[]
    def add_level(self,floor):
        self.level.append(floor)
    def process_entry(self,ticket):
            for level in self.level:
                if level.spots[ticket.veh.type][spottype.free]:
                    ticket.spot =level.assignSpot(ticket)
                    print('Entry Completed for : ',ticket.veh.num)
                    break
    def process_exit(self,ticket):
        for level in self.level :
            if ticket.spot in level.spots[ticket.veh.type][spottype.occupied] :
                level.unassignSpot(ticket)
                break
        ticket.outTime  = time.time()
        ticket.spots    = None
        ticket.status   = ticket_stat.complete
        ticket.payment  = Payment(ticket.inTime, ticket.outTime)
        print('Exit Completed For  : ', ticket.veh.num, ' Pay : ', int(ticket.payment.amount))
class parking_level():
    def __init__(self, name) :
        self.name = name
        self.spots = {vehicletype.four_vehiler : {spottype.free : [], spottype.occupied : []}, 
                        vehicletype.two_vehiler : {spottype.free : [], spottype.occupied : []} }
    def assignSpot(self, ticket) :
        if self.spots[ticket.veh.type][spottype.free] != [] :
            spot = self.spots[ticket.veh.type][spottype.free].pop()
            ticket.spot = spot
            self.spots[ticket.veh.type][spottype.occupied].append(spot)
            return ticket.spot
        return False
    def unassignSpot(self, ticket) :
        self.spots[ticket.veh.type][spottype.free].append(ticket.spot)
        self.spots[ticket.veh.type][spottype.occupied].remove(ticket.spot)
    def addSpot(self, type, num) :
        for eachnum in range(num) :
            spot = Spot(type)
            self.spots[type][spottype.free].append(spot) 
class Vehicle :
    def __init__(self, num) :
        self.id     = self.generateID()
        self.num    = num
    def generateID(self) :
        yield range(100)

class four_vehiler (Vehicle) :
    def __init__(self, num) :
        super().__init__(num)
        self.type   = vehicletype.four_vehiler

class two_vehiler (Vehicle) :
    def __init__(self, num) :
        super().__init__(num)
        self.type   = vehicletype.two_vehiler

class Spot :
    def __init__(self, type) :
        self.id     = self.generateID()
        self.type   = type
    def generateID(self) :
        yield range(100)

class Payment :
    def __init__(self, inTime, outTime) :
        self.mode   = None
        self.rate   = [30, 20, 10]
        self.amount = self.calAmount(inTime, outTime)
    def getRate(self) :
        return self.rate
    def setRate(self, rate) :
        self.rate = rate
    def calAmount(self, inTime, outTime) :
        amount =  (outTime - inTime) * self.getRate()[0]
        amount += (outTime - inTime - 60 ) * self.getRate()[1] if outTime - inTime - 60 > 0 else 0
        amount += (outTime - inTime - 120 ) * self.getRate()[2] if outTime - inTime - 120 > 0 else 0
        return amount        

class ticket :
    def __init__(self, veh) :
        self.veh     = veh
        self.status  = ticket_stat.active
        self.inTime  = time.time()
        self.outTime = None
        self.payment = None
        self.spot    = None
    def generateID(self) :
        # some ID generation mechanism
        return ID

class DisplayBoard :
    def show(self, P) :
        for level in P.level :
            print(P.name , '-' , level.name, '- Available Parking Spots')
            print('Car  : ', len(level.spots[vehicletype.four_vehiler][spottype.free]))
            print('Bike : ', len(level.spots[vehicletype.two_vehiler][spottype.free]))
    
```
